```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import urllib.request
import pytesseract
from pytesseract import Output
import cv2
import numpy as np
import pandas as pd

def img_results(link):
  img = cv2.imread(link)
  #img.shape
  grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  x, y = grayscale.shape
  temp = np.ones((x+10, y),np.uint8)
  temp[10:, :] = grayscale
  _, thresholded = cv2.threshold(temp, 0, 255, cv2.THRESH_BINARY)
  cv2.imwrite("thresholded.png", thresholded)

  results = pytesseract.image_to_string(thresholded, config='--psm 6 --oem 3')
  print(results)
  return results


chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
# options = Options()
# options.headless = True
# driver = webdriver.Chrome('C:/Users/mz/Downloads/chromedriver_win32/chromedriver', options=options)

df = pd.DataFrame({"phone":[], "email":[], "website":[]})


def get_links(link):
    driver.get(link)
    print(link)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    
    box = soup.find('div', {'class': 'bx1'}).find_all('a', href=True)
    print(len(box))
    #card = box.find_all()
    if(len(box) == 0):
        print("END PAGE")
        #email and phone are here
        phone = soup.find('img', {'class', 'sef1'})
        if phone is not None:
            phone = phone['src']
            full_phone_url = "https://www.hotelcontact.net/" + phone
            urllib.request.urlretrieve(full_phone_url, "phone.png")
            phone = img_results('/content/phone.png')

        email = soup.find('img', {'class', 'sef3'})
        if email is not None:
            email = email['src']
            full_email_url = "https://www.hotelcontact.net/" + email
            urllib.request.urlretrieve(full_email_url, "email.png")
            email = img_results('/content/email.png')

        website = soup.find('img', {'class', 'sef2'})
        if website is not None:
            website = website['src']
            full_website_url = "https://www.hotelcontact.net/" + website
            urllib.request.urlretrieve(full_website_url, "website.png")
            website = img_results('/content/website.png')

        # print(phone)
        # print(email)
        df.loc[len(df.index)] = [phone, email, website]
        return
    else:
        for a in soup.find('div', {'class': 'bx1'}).find_all('a', href=True):
            link = a['href']
            get_links(link)
            
        
    
link = "https://www.hotelcontact.net"
get_links(link)

df.to_csv("hotel_data.csv")
```
